```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
data_names=['sai','kamal','venkat','nivas','shiva']
sub1=[45,89,37,20,65]
sub2=[90,68,90,30,20]
sub3=[73,80,45,61,38]
```

```python
plt.bar(data_names,sub1)
```

```python


plt.bar(data_names,sub1)
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)

```

```python

plt.bar(data_names,sub1,color='w',width=0.3,align='edge',edgecolor='y')
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)
```

```python

plt.bar(data_names,sub1,color='w',width=0.6,align='edge',
        edgecolor='y')
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)
```

```python
#linewidth=10,15

plt.bar(data_names,sub1,color='w',width=0.6,align='edge',
        edgecolor='y',linewidth=5)
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)

```

```python
#alpha=0.3,0.5,0.9

plt.bar(data_names,sub1,color='y',width=0.6,align='edge',
        edgecolor='b',linewidth=5,alpha=0.4)
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)

```

```python
#linestyle='--'

plt.bar(data_names,sub1,color='y',width=0.6,align='edge',
        edgecolor='b',linewidth=5,alpha=0.4,linestyle='--')
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend('marks',fontsize=5,loc=3)

```

```python
#visible=True,False

plt.bar(data_names,sub1,color='y',width=0.6,align='edge',
        edgecolor='b',linewidth=5,alpha=0.4,linestyle='--',label='student_marks',visible=True)
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(fontsize=10,loc=1)

```

```python
#visible=True,False

plt.bar(data_names,sub1,color='y',width=0.6,align='edge',
        edgecolor='b',linewidth=5,alpha=0.4,linestyle='--',label='student_marks',visible=False)
plt.title('bar_',fontsize=20)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(fontsize=10,loc=1)

```

```python

plt.figure(figsize=(16,8))

class_index=np.arange(len(data_names))
width=0.4

plt.bar(class_index,sub1,color='y',width=0.4,label='sub1')
plt.bar(class_index+width,sub2,color='r',width=0.4,label='sub2') 

plt.xticks(class_index+width,data_names,rotation=40)
plt.title('student marks',fontsize=20)
plt.xlabel("Students",fontsize=15)
plt.ylabel("marks",fontsize=15)
plt.legend(fontsize=5,loc=3)
 
```

```python
#horizontal graphs

from matplotlib import pyplot as plt

plt.figure(figsize=(10,4))
p1=['kamal','sai','venkat','shiva','santhu','yogi','nivas']
s1=[78,20,102,67,10,43,26]
s2=[54,33,65,23,76,90,84]
s3=[79,70,85,100,93,73,58]

w=0.2

w1=np.arange(len(p1))
#w2=[x+w for x in na]

plt.barh(w1,s1,w)
plt.barh(w1+w,s2,w)
plt.barh(w1+w+w,s3,w)

plt.yticks(w1+w/2,p1,rotation=10)
plt.legend(['s1marks','s2marks','s3marks'])

plt.ylabel('student names')
plt.xlabel('marks')

plt.title('sem 1 marks')
```
