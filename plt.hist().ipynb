```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
age1=np.random.randint(14,50,(25))
age2=np.random.randint(10,40,(25))
```

```python
print(age1)
```

```python
plt.hist(age1)
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python
ins=[10,20,30,40,50]
```

```python
plt.hist(age1,ins)
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python
#rwidth=...    for the area to occupy
```

```python

ins=[10,20,30,40,50]
plt.hist(age1,ins,rwidth=0.9)
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python
#histtype='bar' , 'barstacked', 'step', 'stepfilled'

ins=[10,20,30,40,50]
plt.hist(age1,ins,rwidth=0.9,histtype='step')
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python
#allign

ins=[10,20,30,40,50]
plt.hist(age1,ins,rwidth=0.9,histtype='bar',align='left')
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python


ins=[10,20,30,40,50]
plt.hist(age1,ins,rwidth=0.9,histtype='bar',align='right')
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.show()
```

```python
#orientation

ins=[10,20,30,40,50]
plt.hist(age1,ins,rwidth=0.9,histtype='bar',
         align='left',orientation="horizontal"
        ,color='m',label='student_bars')
plt.title("students_hist")
plt.xlabel("ages")
plt.ylabel("no. of students")
plt.legend(loc=0)
plt.show()
```

```python
age2
```

```python
plt.hist(age1,ins,color='m',rwidth=0.9,label='java_st_ages',
         orientation='vertical',histtype='stepfilled')
plt.hist(age2,ins,color='k',rwidth=0.8,label='py_st_ages',
        orientation='vertical',histtype='bar')
plt.title('java and python students count(ages)',fontsize=20)
plt.xlabel('ages',fontsize=10)
plt.ylabel('no. of students',fontsize=13)
plt.legend(fontsize=9)
plt.show()
```

```python
#representing both the graphs in a single one side by side
```

```python
plt.hist([age1,age2],ins,color=['m','k'],rwidth=0.9,
         label=['java_st_ages','py_st_ages'],
         orientation='vertical')

plt.title('java and python students count(ages)',fontsize=20)
plt.xlabel('ages',fontsize=10)
plt.ylabel('no. of students',fontsize=13)
plt.legend(fontsize=9)
plt.show()
```

```python
#changing the figure size
```

```python
plt.figure(figsize=(16,9))
plt.hist([age1,age2],ins,color=['m','k'],rwidth=0.9,
         label=['java_st_ages','py_st_ages'],
         orientation='vertical')

plt.title('java and python students count(ages)',fontsize=20)
plt.xlabel('ages',fontsize=10)
plt.ylabel('no. of students',fontsize=13)
plt.legend(fontsize=9)
plt.show()
```
